```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import GPy
```

```python
np.random.seed(0)

sig = 0.01

f = lambda x: np.sin(0.3*np.pi*x)
e = lambda x: np.expand_dims(x, -1)
s = lambda x: np.squeeze(x)

x1 = np.linspace(-3, 3, 100)
y1 = f(x1)+np.random.randn(x1.shape[0])*sig

x2 = np.linspace(-30, 30, 100)
y2 = 0.2*f(x2)+np.random.randn(x2.shape[0])*sig

plt.plot(x2, y2)
plt.plot(x1, y1)
plt.show()

x1 = e(x1)
x2 = e(x2)
y1 = e(y1)
y2 = e(y2)
```

```python
k_rbf = GPy.kern.RBF(1, lengthscale=1.0)
```

```python
def gp_pred(xs_1, xs_2, x, y, k):
    jit = 1e-5*np.eye(x.shape[0])
    mu = k.K(xs_1, x)@np.linalg.solve(k.K(x, x)+sig*np.eye(x.shape[0])+jit, y)
    var = k.K(xs_1, xs_2) - k.K(xs_1, x)@np.linalg.solve(k.K(x, x)+sig*np.eye(x.shape[0])+jit, k.K(x, xs_2))
    return mu, var
```

```python
mu, var = gp_pred(x2, x2, x1, y1, k_rbf)
var = np.diag(var)
plt.fill_between(s(x2), s(mu)+s(var), s(mu)-s(var), facecolor='grey')
plt.plot(x2, mu, c='black')
```

# Two layers

```python
k_rbf_1 = GPy.kern.RBF(1, lengthscale=1.0)
mu_1, var_1 = gp_pred(x2, x2, x2, y2, k_rbf_1)
mu_2, var_2 = gp_pred(x2, x2, x1, y1, k_rbf_1)

var_1 = np.diag(var_1)

plt.fill_between(s(x2), s(mu_1)+s(var_1), s(mu_1)-s(var_1), facecolor='grey')
plt.plot(x2, mu_1, c='black')
```

```python
class NormKernel(object):
    def __init__(self, k):
        self.k= k
    
    def normalise(self, k, x1, x2):
        raw_k = k.K(x1, x2)

        k_x1 = np.tile(e(np.diag(k.K(x1, x1))),[1, x2.shape[0]])
        k_x2 = np.tile(np.expand_dims(np.diag(k.K(x2, x2)), 0),[x1.shape[0],1])

        k = np.divide(raw_k, np.sqrt(np.multiply(k_x1, k_x2)))
        return k
    
    def K(self, x1, x2):
        #return self.k.K(x1, x2)
        return self.normalise(self.k, x1, x2)


    
k_dgp = NormKernel(GPy.kern.Linear(1))
#k_dgp = NormKernel(GPy.kern.RBF(1, lengthscale=0.1))
#k = normalise(k_dgp, mu_1, mu_1)
mat = plt.imshow(k)

mat = plt.imshow(k_dgp.K(mu_1, mu_1))
plt.colorbar(mat)
plt.show()
```

# V1

```python
def kk(xs_1, xs_2):
    var_1 = k_rbf.K(xs_1, xs_2)   

    mu_2_xs_1, var_2 = gp_pred(xs_1, xs_2, x2, y2, k_rbf_1)
    mu_2_xs_2, _ = gp_pred(xs_2, xs_1, x2, y2, k_rbf_1)

    return (k_dgp.K(mu_2_xs_1, mu_2_xs_2)+var_1)/2
    #return np.multiply(k_dgp.K(mu_2_xs_1, mu_2_xs_2),var_1)
    #return k_dgp.K(mu_2_xs_1, mu_2_xs_2)
    #return var_1
    #return np.multiply(k_dgp.K(mu_2_xs_1, mu_2_xs_2), GPy.kern.RBF(1, lengthscale=5.0).K(xs_1, xs_2))
```

```python
mat = plt.imshow(kk(x2, x2))
plt.colorbar(mat)
plt.show()
```

```python
xs = x2
mm = kk(xs, x1)@ np.linalg.solve(kk(x1, x1)+sig*np.eye(x1.shape[0]), y1)
#mm = kk(x2, x1)@ y1

plt.plot(x2, y2, label='observed 2', c='black')
plt.scatter(x1, y1, label='observed 1', c='black')

plt.plot(xs, mm, label='DGP Style GP', linewidth=3.0)
plt.plot(xs, k_rbf.K(xs, x1)@np.linalg.solve(k_rbf.K(x1, x1)+sig*np.eye(x1.shape[0]), y1), label='Single GP')

plt.legend()
```

```python
def kk(xs_1, xs_2):
    #y1 = np.ones(x1.shape)
    mu_1, var_1_xs_1 = gp_pred(xs_1, xs_1, x1, y1, k_rbf) #use predictive variance
    mu_1, var_1_xs_2 = gp_pred(xs_2, xs_2, x1, y1, k_rbf) #use predictive variance
    
    var_1_xs_1 = np.diag(np.diag(var_1_xs_1))
    var_1_xs_2 = np.diag(np.diag(var_1_xs_2))
    
    jit = 1e-5
    
    inv_var_1_xs_1 = np.diag(1+jit-np.diag(var_1_xs_1))
    inv_var_1_xs_2 = np.diag(1+jit-np.diag(var_1_xs_2))
    
    mu_2_xs_1, _ = gp_pred(xs_1, xs_1, x2, y2, k_rbf_1)
    mu_2_xs_2, _ = gp_pred(xs_2, xs_2, x2, y2, k_rbf_1)
    #plt.plot(mu_2_xs_1)
    #plt.show()
        

    #return (var_1_xs_1)@k_dgp.K(mu_2_xs_1, mu_2_xs_2)@(var_1_xs_2)+(inv_var_1_xs_1)@k_rbf.K(xs_1, xs_2)@(inv_var_1_xs_2)
    
    kk_dgp = k_dgp.K(mu_2_xs_1, mu_2_xs_2)
    k_xx = k_rbf.K(xs_1, xs_2)
    k = (var_1_xs_1)@kk_dgp@(inv_var_1_xs_2) 
    k = k + (inv_var_1_xs_1)@kk_dgp@(var_1_xs_2)
    k = k + (var_1_xs_1)@kk_dgp@(var_1_xs_2) 
    
    #k = k + k_rbf.K(xs_1, xs_2)
    k = k + 0.5*(inv_var_1_xs_1)@k_xx@(inv_var_1_xs_2)
    k = k + 0.5*(inv_var_1_xs_1)@kk_dgp@(inv_var_1_xs_2)
    return k

    #k = k + k_rbf.K(xs_1, xs_2)
    
    #return k
    #return (var_1_xs_1)@k_rbf.K(xs_1, xs_2)@(var_1_xs_2)+k_dgp.K(mu_2_xs_1, mu_2_xs_2)


    #return var_1
    #return k_xx
    return kk_dgp+k_xx
```
